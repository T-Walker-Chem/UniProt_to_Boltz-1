```python
def create_boltz1_input_for_pair(uniprot1, uniprot2, output_filename, start_entity=1):
    # Fetch sequences and PTMs for both proteins
    sequence1, ptms1 = fetch_uniprot_data(uniprot1)
    sequence2, ptms2 = fetch_uniprot_data(uniprot2)

    gly_constraints = []
    entities = start_entity
    input_content = "version: 1\nsequences:\n"

    # Handle protein 1
    input_content += "  - protein:\n"
    input_content += f"      id: {entity_to_id[entities]}\n"
    input_content += f"      sequence: {sequence1}\n"
    input_content += f"      msa: /lustre/orion/proj-shared/syb111/Personal/twalker/Combined_Pipeline/Sequence_Alignments/{uniprot1}.a3m\n"
    entity_for_mod1 = entities
    entities += 1
    print(entity_to_id[entities])


    # Add modifications (only phosphorylation and glycosylation for now)
    def add_phosphorylation(ptms, entity_id):
        nonlocal entities, input_content, gly_constraints
        # Phosphorylation
        if ptms['phosphorylation']:
            input_content += "    modifications:\n"
            for mod in ptms['phosphorylation']:
                input_content += f"      - position: {mod['position']}\n"
                input_content += f"        ccd: {ptm_to_ccd['phosphorylation'][mod['modification']]}\n"
    
    def add_glycosylation(ptms, entity_id):
        nonlocal entities, input_content, gly_constraints
        # Glycosylation (and constraints)
        if ptms['glycosylation']:
            for mod in ptms['glycosylation']:
                if "N-linked (GlcNAc" in mod['description']:
                    input_content += "  - ligand:\n"
                    input_content += f"      id: {entity_to_id[entities]}\n"
                    input_content += "      ccd: NAG\n"
                    gly_constraints.append([entity_to_id[entity_id], mod['position']['value'], 'N'])
                    gly_constraints.append([entity_to_id[entities], 1, 'C1'])
                    entities += 1
                elif "O-linked (GlcNAc" in mod['description']:
                    input_content += "  - ligand:\n"
                    input_content += f"      id: {entity_to_id[entities]}\n"
                    input_content += "      ccd: NAG\n"
                    gly_constraints.append([entity_to_id[entity_id], mod['position']['value'], 'O'])
                    gly_constraints.append([entity_to_id[entities], 1, 'C1'])
                    entities += 1
                else:
                    continue

    add_phosphorylation(ptms1, entity_for_mod1)
    
    input_content += "  - protein:\n"
    input_content += f"      id: {entity_to_id[entities]}\n"
    input_content += f"      sequence: {sequence2}\n"
    input_content += f"      msa: /lustre/orion/proj-shared/syb111/Personal/twalker/Combined_Pipeline/Sequence_Alignments/{uniprot2}.a3m\n"
    entity_for_mod2 = entities
    entities += 1
    
    add_phosphorylation(ptms2, entity_for_mod2)
    add_glycosylation(ptms1, entity_for_mod1)
    add_glycosylation(ptms2, entity_for_mod2)

    # Add constraints (for glycosylation only here)
    if gly_constraints:
        input_content += "constraints:\n"
        for i in range(0, len(gly_constraints), 2):
            try:
                atom1 = gly_constraints[i]
                atom2 = gly_constraints[i+1]
                input_content += "  - bond:\n"
                input_content += f"      atom1: [{', '.join(map(str, atom1))}]\n"
                input_content += f"      atom2: [{', '.join(map(str, atom2))}]\n"
            except IndexError:
                continue 

    with open(output_filename, 'w') as f:
        f.write(input_content)

    print(f" Boltz-1 input for proteins {uniprot1} & {uniprot2} written to: {output_filename}")

```

```python
from itertools import product

def file_to_list(file_path):
    lines = []
    try:
        with open(file_path, 'r') as file:
            for line in file:
                lines.append(line.rstrip('\n'))
    except FileNotFoundError:
        print(f"Error: File not found: {file_path}")
    return lines

def find_intersection(list1, list2):
    """
    Finds the intersection of two lists.

    Args:
        list1: The first list.
        list2: The second list.

    Returns:
        A new list containing the common elements present in both input lists.
    """
    return list(set(list1) & set(list2))

def unique_combinations(list1, list2):
    combinations_set = set()
    for element1, element2 in product(list1, list2):
        sorted_tuple = tuple(sorted((element1, element2)))
        if element1 != element2:
            combinations_set.add(sorted_tuple)
    return list(combinations_set)


clade_path="/Users/ntw/Desktop/OA_Clade_16/Protein_list.txt"
all_OA_path="/Users/ntw/Desktop/OA_related_genes_original/All_OA_proteins.txt"
cladeToAllOutfile="/Users/ntw/Desktop/OA_Clade_16/Intersection_all_to_clade.txt"
allToCladeOutfile="/Users/ntw/Desktop/OA_Clade_16/Intersection_clade_to_all.txt"
combinations_out="/Users/ntw/Desktop/OA_Clade_16/unique_combinations.txt"

protList = file_to_list(all_OA_path)
cladeList = file_to_list(clade_path)

#print(protList)
#print(cladeList)

intersection_result = find_intersection(protList, cladeList)
with open(cladeToAllOutfile, 'w') as file:
    for i in range(len(intersection_result)):
        file.write(f'{intersection_result[i]}\n')
#print(intersection_result)
print(len(intersection_result))

intersection_result = find_intersection(cladeList, protList)
with open(allToCladeOutfile, 'w') as file:
    for i in range(len(intersection_result)):
        file.write(f'{intersection_result[i]}\n')
#print(intersection_result)
print(len(intersection_result))

combined_list=unique_combinations(intersection_result, intersection_result)
print(combined_list)
with open(combinations_out, 'w') as file:
    for i in range(len(combined_list)):
        file.write(f'{combined_list[i]}\n')

#print(len(combined_list))
#for i in range(len(combined_list)):
    #print(combined_list[i][0])
    #print(combined_list[i][1])



output_prefix = '/Users/ntw/Desktop/OA_Clade_16/'
suffix= '.yaml'

#file_path = output_prefix + uniProtIDs
#print(my_list)

for i in range(len(combined_list)):
    uniprot_1 = (combined_list[i][0])
    uniprot_2 = (combined_list[i][1])
    
    output_filename = output_prefix + uniprot_1 + "_" + uniprot_2 + "/job" + suffix
    output_directory = output_prefix + uniprot_1 + "_" + uniprot_2 + "/"
    fasta_outfile = output_directory +"job.fasta"
    print(output_filename)
    
    #os.makedirs(output_directory, exist_ok=True)
    #entities=0
    #create_boltz1_input_for_pair(uniprot_1, uniprot_2, output_filename, start_entity=1)
    
```
